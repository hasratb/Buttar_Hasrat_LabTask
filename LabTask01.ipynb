### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})
df
```

__Explanation__ For (1) I utilized map to map the female sex to 0 and male sex to 1. This iterates through each element in the Sex column and categorizes it.

```python
def age_category(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(age_category)

df
```

__Explanation__ For (2) I first created the function age_category where I use a series of if statements to determine what category the age input fits in. I also included .isna as an if statement because you cannot compare missing numbers. Then I applied this funciton to every element in the Age column.
